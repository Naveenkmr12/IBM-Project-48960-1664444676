```python
Q1= df.Balance.quantile(0.25)
Q3=df.Balance.quantile(0.75)
```

```python
IQR=Q3-Q1
```

```python
upper_limit =Q3 + 1.5*IQR
lower_limit =Q1 - 1.5*IQR
```

```python
df['Balance'] = np.where(df['Balance']>upper_limit,9,df['Balance'])
```
